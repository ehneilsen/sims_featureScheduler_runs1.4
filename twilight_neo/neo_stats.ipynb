```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import sqlite3
import pandas as pd
```

```python
# database available at:  https://lsst-web.ncsa.illinois.edu/sim-data/sims_featureScheduler_runs1.4/twilight_neo/
conn = sqlite3.connect('twilight_neo_mod1_v1.4_10yrs.db')
df = pd.read_sql('select altitude,sunAlt,visitExposureTime,night,filter,slewTime, observationStartMJD, observationId from SummaryAllProps where note="twilight_neo";', conn)

conn.close()
```

```python
print('exposure times (s) = ', np.unique(df['visitExposureTime']))
print('median altitude of twilight observations=', np.median(df['altitude']))
```

```python
ack = plt.hist(df['altitude'], bins=50)
plt.xlabel('altitude (degrees)')
plt.ylabel('# of obs')
```

```python
ack = plt.hist(df['sunAlt'], bins=50)
plt.xlabel('Sun altitude (degrees)')
plt.ylabel('# of obs')
```

```python
good=np.where(df['night'] < 60)
ack = plt.hist(df['sunAlt'].values[good], bins=50)
```
